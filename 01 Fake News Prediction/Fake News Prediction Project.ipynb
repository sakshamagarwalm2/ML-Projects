In this project, we will use library re to read the tokens We will use nltk corpus stopwords to ignore the words like And him, etc We also include Porter Stream to find the root word And in this project, we will use knn model.
Stamming is the process of reducing a word to its root word.


```python
import numpy as np
import pandas as pd
import re
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
import nltk
nltk.download('stopwords')
```

```python
ds = pd.read_csv('/content/train.csv')
ds.shape
```

```python
ds.isnull().sum()
```

```python
ds = ds.fillna('')
```

```python
#merging titel and author col
ds['content'] = ds['author']+''+ds['title']
```

```python
ps = PorterStemmer()
def stemming(content):
  stemmed_con = re.sub('[^a-zA-Z]',' ', content)
  stemmed_con = stemmed_con.lower()
  stemmed_con = stemmed_con.split()
  stemmed_con = [ps.stem(word) for word in stemmed_con if not word in stopwords.words('english')]
  stemmed_con = ' '.join(stemmed_con)
  return stemmed_con
```

```python
ds['content'] = ds['content'].apply(stemming)
```

```python
# sepreting the data and label
X = ds['content'].values
y = ds['label'].values

# converting the textual data to numerical data
vectorizer = TfidfVectorizer()
vectorizer.fit(X)

X = vectorizer.transform(X)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, y, test_size = 0.2, stratify=y, random_state=2)
model = KNeighborsClassifier()
model.fit(X_train, Y_train)
```

```python
# accuracy score on the test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
test_data_accuracy
```

```python
clf = LogisticRegression()
clf.fit(X_train, Y_train)
X_test_prediction = clf.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
test_data_accuracy
```
