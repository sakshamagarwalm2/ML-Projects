#Whatsapp Chat Analysis

**Preprossesing**

```python
import re
import pandas as pd
```

```python
f = open('/content/WhatsApp Chat with ++Sophomores of MUJ.txt','r',encoding='utf-8')
```

```python
data = f.read()
print(data)
```

```python
print(type(data))
```

```python
pattern = '\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{2}\s-\s'
```

```python
msg = re.split(pattern,data)[1:]
msg
```

```python
dates=re.findall(pattern,data)
dates
```

```python
df = pd.DataFrame({'user_msg':msg,'message_date':dates})
# convert msd_date type
df['message_date'] = pd.to_datetime(df['message_date'],format='%d/%m/%Y, %H:%M - ')
df.rename(columns={'message_date':'date'},inplace=True)

df.head()
```

```python
df.shape
```

```python
# seprate users and msg
user=[]
msg=[]
for i in df['user_msg']:
  entry = re.split('([\w\W]+?):\s',i)
  if entry[1:]:
    user.append(entry[1])
    msg.append(entry[2])
  else:
    user.append('group_notification')
    msg.append(entry[0])
df['user']=user
df['msg']=msg
df.drop(columns=['user_msg'],inplace=True)
df.head()
```

```python
df['year']=df['date'].dt.year
df['month']=df['date'].dt.month_name()
df['day']=df['date'].dt.day
df['hour']=df['date'].dt.hour
df['minute']=df['date'].dt.minute
df.head()
```
